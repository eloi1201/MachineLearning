```
import tensorflow as tf 
import pandas as pd 
```

```
boston_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston_csv = pd.read_csv(boston_url)
boston_csv.head()
```

```
feature_df = boston_csv.drop(['medv'], axis='columns')
feature_df.head()
```

```
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
model.fit(feature_df, boston_csv['medv'], epochs=10000, verbose=0)
model.fit(feature_df, boston_csv['medv'], epochs=10)
```

```
print(model.predict(feature_df[5:10]))
```

```
print(boston_csv['medv'][5:10])
```

```
print(model.get_weights())
```

```
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model_m = tf.keras.models.Model(X,Y)
model_m.compile(loss='mse')
```

```
model_m.summary()
```

```
model_m.fit(feature_df, boston_csv['medv'], epochs=10000, verbose=0)
model_m.fit(feature_df, boston_csv['medv'], epochs=10)
```

```
print(model_m.predict(feature_df[5:10]))
print(boston_csv['medv'][5:10])
```

```
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model_m2 = tf.keras.models.Model(X,Y)
model_m2.compile(loss='mse')
```

```
model_m2.fit(feature_df, boston_csv['medv'], epochs=10000, verbose=0)
model_m2.fit(feature_df, boston_csv['medv'], epochs=10)
```

```
print(model_m2.predict(feature_df[5:10]))
print(boston_csv['medv'][5:10])
```
