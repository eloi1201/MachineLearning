```
import tensorflow as tf
import pandas as pd
```

```
iris_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris_csv = pd.read_csv(iris_url)
iris_csv.head()
```

```
iris_csv = iris_csv.rename(columns={'꽃잎길이':'sepal length','꽃잎폭':'sepal width','꽃받침길이':'petal length','꽃받침폭':'petal width','품종':'label'})
iris_csv.head()
```

```
iris_df = pd.get_dummies(iris_csv)
iris_df.head()
```

```
iris_feature = iris_df[['sepal length','sepal width','petal length','petal width']]
iris_label = iris_df[['label_setosa','label_versicolor','label_virginica']]
```

```
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
model.fit(iris_feature, iris_label, epochs=10000, verbose=0)
model.fit(iris_feature, iris_label, epochs=10)
```

```
print(model.predict(iris_feature[:5]))
print(iris_label[:5])
```

```
print(model.get_weights())
```
