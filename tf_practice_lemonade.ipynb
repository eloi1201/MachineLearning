```
import tensorflow as tf
import pandas as pd
```

```
lemonade_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemonade_csv = pd.read_csv(lemonade_url)
lemonade_csv.head()
```

```
lemonade_df = lemonade_csv.rename(columns={'온도':'degree','판매량':'sell'})
lemonade_df.head()
```

```
X = tf.keras.layers.Input(shape=[1]) # The number of feature is 1
Y = tf.keras.layers.Dense(1)(X)      # The number of label is 1
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')            # Define the way of training
```

```
# The value in epochs variable means how many times the data will be learned repeatedly. Loss value will be reducted through repeated learning.
model.fit(lemonade_df['degree'], lemonade_df['sell'], epochs=10000, verbose=0)
model.fit(lemonade_df['degree'], lemonade_df['sell'], epochs=10)
```

```
print(model.predict([[15]]))
```

```
print(model.get_weights())
```
